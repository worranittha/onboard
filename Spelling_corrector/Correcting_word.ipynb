# Find unique words in all reports

```python
import re

unique_word = set()
for report in df['Reports']:
    for sent in report.split('\n'):
        sent = (re.sub('[.,;:=\\-()\[\]{}<>/+?]', ' ', sent)).strip()
        for word in sent.split():
            word = word.strip().lower()
            if word != '' and not re.search('\d', word) and len(word) > 3:
                unique_word.add(word)

print('len unique word :', len(unique_word))
unique_word
```

# Using pattern

pattern แก้ dict ยาก ไม่มี resource เลยเลือกใช้อย่างอื่นแทน

Import library

```python
from pattern.en import suggest
```

Example

```python
# suggest from pattern
print('suggest :', suggest('efusion'))
print('first suggest :', suggest('efusion')[0][0])
```

Correcting unknown words (use unknown keywords from pyspellchecker)

```python
def correctPattern(word, unknown_key_spell):
    for w in unknown_key_spell:
        if nltk.edit_distance(word, w) <= 0.2*len(w):
            return w
    return suggest(word)[0][0]
```

```python
correct_pattern = {word: correctPattern(word, unknown_key_spell) for word in unique_word}
correct_pattern
```

Save correct word to json file

```python
import json
with open("Correct_word_pattern.json", "w") as fp:
    json.dump(correct_pattern,fp) 
```

# Using pyspellchecker

Import library

```python
from spellchecker import SpellChecker
```

Example

```python
# initialize the instance
spell = SpellChecker()

# check if words are found in the dictionary
print('known :', spell.known(['cardiomegaly', 'opacities', 'densities']))
# check if words are not found in the dictionary
print('unknown :', spell.unknown(['cardiomegaly', 'opacities', 'densities']))

# add new word to existing dictionary
# spell.word_frequency.load_words(['cardiomegaly']) # add many words
spell.word_frequency.add('cardiomegaly') # add single word

# find those words that may be misspelled
misspelled = spell.unknown(['cardioemgaly','something'])
for word in misspelled:
    # get the one `most likely` answer
    print(spell.correction(word))
    # get a list of `likely` options
    print(spell.candidates(word))
```

```python
spell = SpellChecker()
```

# Using symspellpy

Import library

```python
from symspellpy import SymSpell, Verbosity
import pkg_resources
from itertools import islice
```

Example

```python
# initiallize and load dictionary
sym_spell = SymSpell()

# single word
dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)
# term_index is the column of the term and count_index is the
# column of the term frequency
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
print(list(islice(sym_spell.words.items(), 5)))

# bigram
sym_spell_2 = SymSpell()
dictionary_path_2 = pkg_resources.resource_filename(
    "symspellpy", "frequency_bigramdictionary_en_243_342.txt"
)
sym_spell_2.load_bigram_dictionary(dictionary_path_2, 0, 2)
print(list(islice(sym_spell_2.bigrams.items(), 5)))
```

```python
# lookup suggestions for single-word input strings
input_term = "memebers"  # misspelling of "members"
# max edit distance per lookup
# (max_edit_distance_lookup <= max_dictionary_edit_distance)
suggestions = sym_spell.lookup(input_term, Verbosity.CLOSEST, max_edit_distance=2, include_unknown=True)
# display suggestion term, edit distance, and term frequency
for suggestion in suggestions:
    print(suggestion)

# print(suggestions)
```

Initialize symspell and load dictionary

```python
sym_spell = SymSpell()
dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
```

Find keywords that are not in symspell dictionary and add them to dictionary

```python
known_word_sym = [w for w in unique_word if w in sym_spell.words.keys()]
unknown_word_sym = [w for w in unique_word if w not in sym_spell.words.keys()]

print(len(known_word_sym))
print('Example if cardiomegaly in dict :','cardiomegaly' in known_word_sym)
```

```python
unknown_key_sym = [w for w in all_keywords if w not in known_word_sym]
unknown_key_sym
```

```python
for w in unknown_key_sym:
    sym_spell.create_dictionary_entry(w, 999999999)

# test new dict
print('Example if costophrenic in dict :','costophrenic' in sym_spell.words.keys())
```

Correcting unknown words

```python
def correctSymspell(word, sym_spell):
    return sym_spell.lookup(word, Verbosity.TOP, max_edit_distance=2, include_unknown=True)[0].term
```

```python
correct_sym = {word: correctSymspell(word, sym_spell) for word in unknown_word_sym}
correct_sym
```

Save correct words to json file

```python
import json
with open("Correct_word_symspell.json", "w") as fp:
    json.dump(correct_sym,fp) 
```

# Using jamspell

Import library, initialize jamspell load the model

```python
import jamspell

jam_spell = jamspell.TSpellCorrector()
jam_spell.LoadLangModel('en.bin')
```

Example

```python
jam_spell.FixFragment('I am the begt spell cherken!')
# # u'I am the best spell checker!'

jam_spell.GetCandidates(['i', 'am', 'the', 'begt', 'spell', 'cherken'], 3)
# # (u'best', u'beat', u'belt', u'bet', u'bent', ... )

jam_spell.GetCandidates(['i', 'am', 'the', 'begt', 'spell', 'cherken'], 5)
# # (u'checker', u'chicken', u'checked', u'wherein', u'coherent', ...)
```

Correcting unknown words

```python
def correctJamspell(word, jam_spell):
    return jam_spell.FixFragment(word)
```

```python
correct_jamspell = {word: correctJamspell(word, jam_spell) for word in unknown_word_spell}
correct_jamspell
```
