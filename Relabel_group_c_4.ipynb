Correcting misspelled words

```python
import re
from symspellpy import SymSpell, Verbosity
import pkg_resources
import nltk

def uniqueWord(df):
    '''
        Description: Find unique words in all reports.
        Input:
            df (dataframe): A dateframe 
        Output:
            unique_word (set): A set of unique words in all reports 
    '''
    unique_word = set()
    for report in df['Reports']:
        for sent in report.split('\n'):
            sent = (re.sub('[.,;:=\\-()\[\]{}<>/+?]', ' ', sent)).strip()
            for word in sent.split():
                word = word.strip().lower()
                if word != '' and not re.search('\d', word) and len(word) > 3:
                    unique_word.add(word)
    print('len unique word :', len(unique_word))
    return unique_word

def correctWord(word, all_keywords, sym_spell):
    '''
        Description: Correct input word.
        Input:
            word (str): A word that we need to correct
            sym_spell (SymSpell): A symspell corrector
        Output:
            correct word
    '''
    for k in all_keywords:
        if nltk.edit_distance(word, k) < 0.3*len(k):
            return k
    return sym_spell.lookup(word, Verbosity.TOP, max_edit_distance=2, include_unknown=True)[0].term

def symspellCorrector(df, all_keywords):
    '''
        Description: Correct misspelled words in reports using symspell
        Intput:
            keyword (list): A list of all keywords that are neccessary in re-label step
        Output:
            correct_word (dict): A dictionary of unique words as key and their correct words as value
    '''
    sym_spell = SymSpell()
    dictionary_path = pkg_resources.resource_filename("symspellpy", "frequency_dictionary_en_82_765.txt")
    # bigram_path = pkg_resources.resource_filename("symspellpy", "frequency_bigramdictionary_en_243_342.txt")
    sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
    # sym_spell.load_bigram_dictionary(bigram_path, term_index=0, count_index=2)

    unique_word = uniqueWord(df)
    known_word = [w for w in unique_word if w in sym_spell.words.keys()]
    unknown_word = [w for w in unique_word if w not in known_word]

    # add custom words
    for w in all_keywords:
        if w not in known_word:
            sym_spell.create_dictionary_entry(w, 999999999) 
    
    correct_word = {word: correctWord(word, all_keywords, sym_spell) for word in unknown_word}
    return sym_spell, correct_word, known_word, unknown_word
```

## Function

### Preprocess Function

```python
# correcting reports in dataframe
import re

def splitWord(word, keyword, known_word, ninja):
    '''
        Description: Split word if there are known continuous words, such as enlargedcardiac (enlarged cardiac)\n
        Input:
            word (str): A string of continuous word\n
            keyword (list): A list of keywords\n
            known_word (list): A list of known words\n
            ninja (model): A language model from wordninja modules\n
        Output:
            A list of words that are already split\n
    '''
    word_seg = ninja.split(word)
    return word_seg if word not in keyword + known_word and all(i in known_word for i in word_seg) else [word]

def appendWord(word, sentence, correct_word, keyword, known_word, ninja):
    '''
        Description: Correcting misspelled words\n
        Input:
            word (str): A string of continuous word\n
            sentence (list): A list of words in current sentence\n
            correct_word (dict): A dictionary of correct words
            keyword (list): A list of keywords\n
            known_word (list): A list of known words\n
            ninja (model): A language model from wordninja modules\n
        Output:
            sentence (list): A list of words in current sentence with new words that are already corrected and split\n
    '''
    word_list = splitWord(word, keyword, known_word, ninja)
    for wordl in word_list:
        sentence.append(correct_word[wordl] if wordl in correct_word.keys() else wordl)
    return sentence

def preprocess(df, correct_word, keyword, known_word, ninja):
    '''
        Description: Preprocess the reports by cleaning unwanted symbols, 
                     correcting misspelled words and separating sentences.\n
                     1) Clean unwanted symbols using re.search(r'[^(\s\d).]{3,}[.][^(\s\d).]{3,}', sentence).\n
                     2) Correct misspelled words using symspell package that already prepared in correct_wprd dictionary.\n
                     3) Separate sentences using '\n'. Then check if there are more than 1 sentence in the same line
                        using re.search(r'.{3,}[.]\Z', word). Finally, add <s> to the beginning of a sentence 
                        and <\s> to the end of a sentnece\n
        Input: 
                df (dataframe): A dataframe of all reports\n
                correct_word (dict): A dictionary of corrected words of misspelled words\n
        Output: 
                new_df (dataframe): A dataframe of all reports that has 'Report List' column with proprocessed reports.\n
    '''
    new_df = df.copy()
    report = []
    sentence = []
    prev_sent = []
    con = ['and', 'or', 'be', 'is', 'are', 'was', 'were']
    for i in range(df['Reports'].shape[0]):
        for sent in (df['Reports'].iloc[i]).split('\n'):
            # clean unwanted symbols
            sent = (re.sub('[,;:=\\-()\[\]{}<>/+?]', ' ', sent)).strip()
            # check if there is 'sentence.sentence' in the same line  
            if re.search(r'[^(\s\d).]{3,}[.]', sent):
                sent = ' '.join([re.sub('[.]', '. ', w) if re.search(r'[^\d.]{2,}[.][^\d.]{2,}', w) else w for w in sent.split()])
            # correct misspelled words and separate sentences
            for w,word in enumerate(sent.split()):
                word = word.strip().lower()
                if word != '':
                    if re.search(r'[^(.).]{3,}[.]\Z', word) or w == len(sent.split()) - 1:
                        word = re.sub(r'[.]\Z', '', word)
                        sentence = appendWord(word, sentence, correct_word, keyword, known_word, ninja)
                        # sentence.append(word)
                        # check if current sentence is in the same previous sentence
                        if len(sentence) > 0 and len(prev_sent) > 0 and (sentence[0] in con or prev_sent[-1] in con):
                                sentence = prev_sent + sentence
                                # correct_sent = sym_spell.lookup_compound(sentence, 2)[0].term
                                # report[-1] = '<s> '+correct_sent.strip()+' <\s>'
                                report[-1] = '<s> '+(' '.join(sentence)).strip()+' <\s>'
                        else:
                            # correct_sent = sym_spell.lookup_compound(sentence, 2)[0].term
                            # report[-1] = '<s> '+correct_sent.strip()+' <\s>'
                            report.append('<s> '+(' '.join(sentence)).strip()+' <\s>')
                        prev_sent = sentence.copy()
                        sentence = []
                    else:
                        sentence = appendWord(word, sentence, correct_word, keyword, known_word, ninja)
                        # sentence.append(word)
        new_df.at[i ,'Report List'] = report
        report = []
        prev_sent = []
    return new_df
```

### Re-label Function

```python
# re-label 
import re

def relabel(name, df, keyword, negkey, poskey, uncerkey):
  '''
    Description: Re-label the preprocessed reports using keyword, negkey, poskey, uncerkey and pattern.
                 Use re.search() for searching pattern that contains keyword, negkey, poskey or uncerkey in a sentence.
                 Moreover, we have to check cardiothoracic ratio, so we need to add pattern for checking the ratio's number.
    Input:
      name (str): An observation name
      df (dataframe): A dataframe of proprocessed reports
      keyword (str): A keyword of the observation
      negkey (str): A negative keyword of the observation that are used to check for negative label (0)
      poskey (str): A positive keyword of the observation that are used to check for positive label (1)
      uncerkey (str): A negative keyword of the observation that are used to check for uncertainty label (-1)
    Output:
    new_df (dataframe): A dataframe with observation's ground truth that contains re-labeled results
  '''
  new_df = df.copy()
  report_list = df['Report List']

  # define verd for using in pattern
  verb = ' (is|are|was|were|has|appear|appears|cannot be|shows) .*'
  if name == 'Cardiomegaly': check_name = 'cardiomegaly|left ventricular enlarge|lvh'
  elif name == 'Pleural Effusion': check_name = '.*pleural effusion'
  else: check_name = keyword
  
  for i,report in enumerate(report_list):
    # re-label only reports that BERT and Inspectra gives different results
    if new_df.at[i, name + ' BERT Labeler'] != new_df.at[i, name + ' Inspectra Labeler']:
      for s in report[1:]:
        if keyword != None and re.search(keyword, s):
          # check uncertainty pattern
          if re.search(('(<s>|the) ')+keyword+verb+uncerkey, s) or re.search(uncerkey+keyword, s):
            new_df.loc[i, name + ' Ground Truth'] = -1
            break
          # check negative pattern
          elif re.search(('(<s>|the) ')+keyword+verb+negkey, s) or re.search(negkey+keyword, s):
            new_df.loc[i, name + ' Ground Truth'] = 0
            break
          # check cardiothoracic ratio
          elif name == 'Cardiomegaly' and re.search(r'ratio.* \d+\.\d+', s):
            n = float(re.findall('\d+\.\d+', s)[0])
            new_df.loc[i, name + ' Ground Truth'] = 0 if n <= 0.5 else 1
            break
          # check positive pattern
          elif re.search(('(<s>|the) ')+keyword+verb, s):
            if re.search(('(<s>|the) ')+keyword+verb+poskey, s):
              new_df.loc[i, name + ' Ground Truth'] = 1
              break
          elif re.search(poskey+keyword, s) or re.search('<s> '+check_name, s) or re.search('there is '+keyword, s):
              new_df.loc[i, name + ' Ground Truth'] = 1
              break 
    else:
      new_df.at[i, name + ' Ground Truth'] = new_df.at[i, name + ' Inspectra Labeler']
  
  diff_df = new_df[new_df[name + ' BERT Labeler'] != new_df[name + ' Inspectra Labeler']]
  
  print('Inspectra 0 :', df[name + ' Inspectra Labeler'].value_counts()[0], 'BERT 0 :', df[name + ' BERT Labeler'].value_counts()[0])
  print('Inspectra 1 :', df[name + ' Inspectra Labeler'].value_counts()[1], 'BERT 1 :', df[name + ' BERT Labeler'].value_counts()[1])
  print('Inspectra -1 :', df[name + ' Inspectra Labeler'].value_counts()[-1] if -1 in df[name + ' Inspectra Labeler'].unique() else 0, 'BERT -1 :', df[name + ' BERT Labeler'].value_counts()[-1] if -1 in df[name + ' BERT Labeler'].unique() else 0)
  
  print('Different label between Inspectra and BERT :', diff_df.shape[0])
  print('Relabel 0 :', diff_df[name + ' Ground Truth'].value_counts()[0] if 0 in diff_df[name + ' Ground Truth'].unique() else 0)
  print('Relabel 1 :', diff_df[name + ' Ground Truth'].value_counts()[1] if 1 in diff_df[name + ' Ground Truth'].unique() else 0)
  print('Relabel -1 :', diff_df[name + ' Ground Truth'].value_counts()[-1] if -1 in diff_df[name + ' Ground Truth'].unique() else 0)
  print('Unlabel :', diff_df[name + ' Ground Truth'].value_counts()[""])
  return new_df
```
