## Function

### Re-label Function

```python
# re-label 
import re

def relabel(name, df, keyword, negkey, poskey, uncerkey):
  newdf = df.copy()
  report_list = df['Report List']
  verb = ' (is|are|was|were|has|appear|appears|be) .*'
  if name == 'Cardiomegaly': check_name = 'cardiomegaly|left ventricular enlarge|lvh'
  elif name == 'Pleural Effusion': check_name = '.*pleural effusion'
  else: check_name = keyword
  
  for i,report in enumerate(report_list):
    if newdf.at[i, name + ' BERT Labeler'] != newdf.at[i, name + ' Inspectra Labeler']:
      for s in report[1:]:
        if keyword != None and re.search(keyword, s):
          if re.search(('(<s>|the) ')+keyword+verb+uncerkey, s) or re.search(uncerkey+keyword, s):
            newdf.loc[i, name + ' Ground Truth'] = -1
            # newdf.loc[i, name + ' Key Detect'] = re.search(uncerkey+keyword, s) if re.search(uncerkey+keyword, s) else re.search(keyword+verb+uncerkey, s)
            break
          elif re.search(('(<s>|the) ')+keyword+verb+negkey, s) or re.search(negkey+keyword, s):
            newdf.loc[i, name + ' Ground Truth'] = 0
            # newdf.loc[i, name + ' Key Detect'] = re.search(negkey+keyword, s) if re.search(negkey+keyword, s) else re.search(keyword+verb+negkey, s)
            break
          elif name == 'Cardiomegaly' and re.search(r'ratio.* \d+\.\d+', s):
            n = float(re.findall('\d+\.\d+', s)[0])
            newdf.loc[i, name + ' Ground Truth'] = 0 if n <= 0.5 else 1
            # newdf.loc[i, name + ' Key Detect'] = n
            break
          elif re.search(('(<s>|the) ')+keyword+verb, s):
            if re.search(('(<s>|the) ')+keyword+verb+poskey, s):
              newdf.loc[i, name + ' Ground Truth'] = 1
              # newdf.loc[i, name + ' Key Detect'] = re.search(keyword+verb+poskey, s)
              break
          elif re.search(poskey+keyword, s) or re.search('<s> '+check_name, s):
              newdf.loc[i, name + ' Ground Truth'] = 1
              # newdf.loc[i, name + ' Key Detect'] = re.search(poskey+keyword, s) if re.search(poskey+keyword, s) else re.search('<s> '+check_name, s)
              break  
    else:
      newdf.at[i, name + ' Ground Truth'] = newdf.at[i, name + ' Inspectra Labeler']
  
  print('Ground truth 0 :', newdf[name + ' Ground Truth'].value_counts()[0] if 0 in newdf[name + ' Ground Truth'].unique() else 0)
  print('Ground truth 1 :', newdf[name + ' Ground Truth'].value_counts()[1] if 1 in newdf[name + ' Ground Truth'].unique() else 0)
  print('Ground truth -1 :', newdf[name + ' Ground Truth'].value_counts()[-1] if -1 in newdf[name + ' Ground Truth'].unique() else 0)
  print('Unlabel :', newdf[name + ' Ground Truth'].value_counts()[""])

  return newdf
```

### Correct Report Function

```python
# correcting reports in dataframe
import re

def correctReport(df, correct_word):
  newDf = df.copy()
  report = []
  sentence = []
  for i in range(df['Reports'].shape[0]):
    for s,sent in enumerate((df['Reports'].iloc[i]).split('\n')):
      sent = (re.sub('[,;:=\\-()\[\]{}<>/+?]', ' ', sent)).strip()
      if re.search(r'[^(\s\d).]{3,}[.][^(\s\d).]{3,}', sent):
        sent = ' '.join([re.sub('[.]', '. ', w) if re.search(r'[^\d][.][^\d]', w) else w for w in sent.split()])
      for w,word in enumerate(sent.split()):
        word = word.strip().lower()
        if word != '':
          if re.search(r'.{3,}[.]\Z', word) or w == len(sent.split()) - 1:
            word = re.sub(r'[.]\Z', '', word)
            sentence.append(correct_word[word] if word in correct_word.keys() else word)
            report.append('<s> '+(' '.join(sentence)).strip()+' <\s>')
            sentence = []
          else:
            sentence.append(correct_word[word] if word in correct_word.keys() else word)
    newDf.at[i ,'Report List'] = report
    report = []

  return newDf
```
